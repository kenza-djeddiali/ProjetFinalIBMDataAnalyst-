```python
api_url="https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/labs/module%201/Accessing%20Data%20Using%20APIs/jobs.json"

def get_number_of_jobs_T(technology):
    #your code goes here
    # return technology,number_of_jobs
    response = requests.get(api_url)
    if response.ok:
        jobs = response.json()
        # On compte combien de fois la technologie est présente dans le champ "Key Skills"
        number_of_jobs = sum(technology.lower() in job['Key Skills'].lower() for job in jobs)
        return technology, number_of_jobs
    else:
        return technology, 0
```

#### In the similar way, you can try for below given technologies and results  can be stored in an excel sheet.


Collect the number of job postings for the following languages using the API:

*   C
*   C#
*   C++
*   Java
*   JavaScript
*   Python
*   Scala
*   Oracle
*   SQL Server
*   MySQL Server
*   PostgreSQL
*   MongoDB


```python
# your code goes here
import requests
from openpyxl import Workbook

technologies = [
    "C", "C#", "C++", "Java", "JavaScript", "Python", "Scala",
    "Oracle", "SQL Server", "MySQL Server", "PostgreSQL", "MongoDB"
]

api_url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/" \
          "IBM-DA0321EN-SkillsNetwork/labs/module%201/Accessing%20Data%20Using%20APIs/jobs.json"

def get_number_of_jobs_T(technology):
    response = requests.get(api_url)
    if response.ok:
        jobs = response.json()
        number_of_jobs = sum(
            technology.lower() in job.get("Key Skills", "").lower()
            for job in jobs
        )
        return technology, number_of_jobs
    else:
        return technology, 0

wb = Workbook()
ws = wb.active
ws.append(["Technology", "Number of Job Postings"])

for tech in technologies:
    tech_name, count = get_number_of_jobs_T(tech)
    print(f"{tech_name}: {count}")
    ws.append([tech_name, count])

wb.save("github-job-postings.xlsx")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Charger les données du fichier Excel
df = pd.read_excel("job-postings.xlsx")

# Trier les données par nombre d'offres d'emploi en ordre décroissant
df_sorted = df.sort_values(by="Number of Job Postings", ascending=False)

# Créer le graphique à barres
plt.figure(figsize=(10, 6))
plt.bar(df_sorted["Technology"], df_sorted["Number of Job Postings"], color="cornflowerblue")

# Ajouter le titre et les étiquettes en français
plt.title("Nombre d'offres d'emploi par technologie", fontsize=14, fontweight='bold')
plt.xlabel("Technologie", fontsize=12)
plt.ylabel("Nombre d'offres d'emploi", fontsize=12)

# Rotation des étiquettes sur l'axe X pour une meilleure lisibilité
plt.xticks(rotation=45, ha="right")

# Ajuster automatiquement la mise en page
plt.tight_layout()

# Afficher le graphique
plt.show()
```
